## 单变量线性回归

```python
train_x = np.linspace(-1,1,100)
```

```python
# 参数
w0 = 3
b0 = 2
```

```python
# 创建y,加入随机噪声
train_y = w0*train_x +b0 + np.random.randn(len(train_x))*0.2
```

```python
plt.plot(train_x, train_y, 'o')
plt.plot(train_x, w0*train_x+b0, linewidth=3)
```

tensorflow 2.0 取消了placeholder

```python
# 定义占位符
x1 = tf.placeholder(tf.float32,name='x')
y1 = tf.placeholder(tf.float32,name='y')
```

```python
# 定义模型函数
def model(x,w,b):
    return tf.multiply(w,x)+b
```

```python
# 定义初始参数及模型结构
wt = tf.Variable(0.2,name='w')
bt = tf.Variable(1.0,name='b')
pred = model(x1,wt,bt)
```

```python
# 定义迭代次数和学习率
train_epochs = 10
learning_rate = 0.1
```

```python
# 均方差作为损失函数
loss_func = tf.reduce_mean(tf.square(y1-pred))
```

```python
# 梯度下降优化最小化损失函数
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss_func)
```

```python
# 创建会话并初始化
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# 迭代拟合
for i in range(train_epochs):
    for xt,yt in zip(train_x,train_y):
        _,loss = sess.run([optimizer,loss_func],feed_dict={x1:xt,y1:yt})
        b0temp = bt.eval(session=sess)
        w0temp = wt.eval(session=sess)
    plt.plot(train_x,train_x*w0temp+b0temp)
```

```python
# 拟合后的斜率和偏置
sess.run(wt),sess.run(bt)
```

```python
plt.plot(train_x,train_y,'o',train_x,w0*train_x +b0,linewidth=3)
plt.plot(train_x,2.8285425*train_x+1.9722569,linewidth=3)
```

```python
# 预测
x_test = 2.3
predict = sess.run(pred,feed_dict={x1:x_test})
```

```python
print("预测值%s"%predict)
print("真实值%s"%(w0*x_test +b0))
```

```python
# 关闭会话
sess.close()
```

## 多元线性回归

```python
# 加载数据
house_price = pd.read_excel('./boston housing price.xlsx')
```

```python
house_price.describe()
```

```python
# 归一化
for col,ser in house_price.iloc[:,:12].iteritems():
    min_t,max_t = ser.min(),ser.max()
    house_price[col] = ser.apply(lambda x:(x-min_t)/(max_t-min_t))
```

```python
# 随机划分训练集与测试集
train_index = np.random.randn(house_price.shape[0])<0.8
train = house_price.loc[train_index]
test = house_price.loc[~train_index]
```

```python
train.shape, test.shape
```

```python
train_x,train_y,test_x,test_y = train.iloc[:,:12].values, train.iloc[:,12].values, test.iloc[:,:12].values, test.iloc[:,12].values
```

```python
# 定义占位符
xt = tf.placeholder(shape=[None,12],dtype=np.float32,name='x')
yt = tf.placeholder(shape=[None,1],dtype=tf.float32,name='y')
```

```python
# 定义参数和模型结构
w0 = tf.Variable(tf.random_normal([12,1],dtype=np.float32,stddev=0.1),name='w') #  初始化w，服从标准差为0.01的随机正态分布的数
b0 = tf.Variable(1.0,dtype=np.float32,name='b')
def model2(x,w,b):
    return tf.matmul(x,w)+b ## 矩阵乘法，w在后，x在前
```

```python
predt = model2(xt,w0,b0)
```

```python
# 创建命名空间，给损失函数命名，定义损失函数,采用均方差作为损失函数
with tf.name_scope("LossFunction"):
    loss_func = tf.reduce_mean(tf.square(yt - predt))
# loss_func = tf.reduce_mean(tf.square(yt-predt))
```

```python
# 迭代次数和步长
train_epochs = 100
learning_rate = 0.05
```

```python
# 梯度下降优化算法
optimizers = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss_func)
```

```python
# 创建会话并初始化变量
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# 训练模型并绘制损失曲线
loss_list = []
for i in range(train_epochs):
    loss_v = 0
    for xs, ys in zip(train_x,train_y):
        xs = xs.reshape(1,12)
        ys = ys.reshape(1,1)
        _,loss = sess.run([optimizers,loss_func],feed_dict={xt:xs,yt:ys})
        loss_v+=loss
    loss_list.append(loss_v/train_x.shape[0])
    w0temp = w0.eval(session=sess)
    b0temp = b0.eval(session=sess)
```

```python
# 每轮迭代代价下降曲线
plt.plot(loss_list)
```

```python
# 测试集预测
test_pred = sess.run(predt,feed_dict={xt:test_x}).reshape(test_x.shape[0])
```

```python
# 训练集预测
train_pred = sess.run(predt,feed_dict={xt:train_x}).reshape(train_x.shape[0])
```

### 单神经元

```python
# 创建占位符
xp = tf.placeholder(dtype=tf.float32,shape=[None,784],name='x')
yp = tf.placeholder(dtype=tf.float32,shape=[None,10],name='y')
```

```python
# 定义参数和模型
w = tf.Variable(tf.zeros([784,10]),name='w')
b = tf.Variable(tf.zeros([10]),name='b')
```

```python
# 模型结构
y_ = tf.nn.softmax(tf.matmul(xp,w)+b)
```

```python
train_epochs=1000 # 迭代次数
learning_rate = 0.01 # 这里学习率不能设太大
```

```python
# 1. 交叉熵损失
with tf.name_scope("CrossEntropy"):
    cross_entropy = -tf.reduce_sum(yp*tf.log(y_))
# # 2.对数损失损失
# with tf.name_scope("LossFunction"):
#     loss_function = tf.reduce_mean(-tf.reduce_sum(yp * tf.log(y_) + (1 - yp) * tf.log(1 - y_), reduction_indices=1))
```

```python
# 梯度下降优化
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cross_entropy)
```

```python
# 创建会话并初始化变量
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# 每轮迭代选取100个样本做优化运算
for i in range(train_epochs):
    xs,ys = mnist.train.next_batch(100)
    sess.run(optimizer,feed_dict={xp:xs,yp:ys})
    w0t = w.eval(session=sess)
    b0t = b.eval(session=sess)
```

```python
# 结果验证1：原始方法
# 测试集预测值
test_pred = np.argmax(sess.run(y_,feed_dict={xp:mnist.test.images}),axis=1)
# 测试集真实值
test_y = np.argmax(mnist.test.labels,axis=1)
```

```python
print("test accuracy is %s"%((test_pred==test_y).sum()/test_y.shape[0]))
```

```python
# 结果验证2：tf方法
correct_prediction = tf.equal(tf.argmax(y_, 1), tf.argmax(yp, 1))
```

```python
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
sess.run(accuracy, feed_dict={xp: mnist.test.images, yp: mnist.test.labels})
```

### 多层神经网络

```python
xp = tf.placeholder(tf.float32,[None,784],'x')
yp = tf.placeholder(tf.float32,[None,10],'y')
```

```python
# 参数定义。2个隐藏层，1个输出层
w1 = tf.Variable(tf.random_normal([784,256],stddev=0.01),'w1')
b1 = tf.Variable(tf.zeros([256]),'b1')

w2 = tf.Variable(tf.random_normal([256,128],stddev=0.01),'w2')
b2 = tf.Variable(tf.zeros([128]),'b2')

w3 = tf.Variable(tf.random_normal([128,10],stddev=0.01),'w3')
b3 = tf.Variable(tf.zeros([10]),'b3')
```

```python
# 模型结构
l1 = tf.nn.relu(tf.matmul(xp,w1)+b1)
l2 = tf.nn.relu(tf.matmul(l1,w2)+b2)
y_ = tf.nn.softmax(tf.matmul(l2,w3)+b3)
```

```python
# 交叉熵损失
with tf.name_scope('CrossEntropy2'):
    cross_entropy = -tf.reduce_sum(yp*tf.log(y_))
```

```python
# 梯度下降优化算法
with tf.name_scope('lossFunc'):
    optimizers = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# 模型训练
for i in range(2000):
    xs,ys = mnist.train.next_batch(100)
    sess.run(optimizers,feed_dict={xp:xs,yp:ys})
```

```python
# 测试集准确性
coll = tf.equal(tf.argmax(y_,1),tf.argmax(yp,1))
accuracy = tf.reduce_mean(tf.cast(coll,tf.float32))
sess.run(accuracy,feed_dict={xp:mnist.test.images, yp:mnist.test.labels})
```

### quick start

```python
# 层的线性堆叠
model = keras.Sequential()
```

```python
# 输入层
model.add(layers.Dense(32, activation='relu'))
# 隐藏层
model.add(layers.Dense(64,activation='relu'))
# 输出层
model.add(layers.Dense(10,activation='softmax'))
```

```python
# 配置模型训练，优化算法，损失函数，评价标准
model.compile(optimizer=keras.optimizers.Adam(0.01), loss=keras.losses.categorical_crossentropy, metrics=[keras.metrics.categorical_accuracy])
```

```python
# 训练模型。
model.fit(train_x, train_y, epochs=10, batch_size=100, validation_data=(val_x, val_y))
```

### 方法及参数

1. 快速手写数字识别

```python
# 加载手写数据集
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
x_train.shape, x_test.shape
```

- keras.layers.Flatten：将数据展平
- keras.layers.Dropout："抓爆"，随机的暂时将部分神经单元丢弃，常用于防止过拟合

```python
# layers参数设置：kernel使用l1正则化，bias使用l2正则化，kernel初始化为随机正交矩阵，bias初始化为常数2.0
model = keras.Sequential(
   [ layers.Flatten(input_shape=(28,28)),
    layers.Dense(128,activation='relu',kernel_regularizer=keras.regularizers.l1(0.01),
                 bias_regularizer=keras.regularizers.l2(0.01),
                 kernel_initializer='orthogonal',bias_initializer=tf.keras.initializers.Constant(2.0)),
    layers.Dropout(rate=0.1),
    layers.Dense(10,activation='softmax')]
)
```

1. optimizer: keras.optimizers中有各类优化算法
2. loss: keras.losses中含损失函数。字符串指定常用默认损失函数如"mse","categorical_crossentropy","binary_crossentropy".
3. metrics： keras.metrics中含度量函数。
4. 设置run_eagerly=True让评估更快(2.0版支持)。

```python
# 编译模型,指定优化算法，损失函数和度量
# optimizer参数：可使用keras.optimizers中优化算法，也可用字符串（如'adam','sgd'）指定默认方法
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

- epochs：迭代次数。
- batch_size：每次迭代使用的样本数
- validation_data：实时监测验证集上效果，每次迭代执行一次

```python
%%time
# 训练
model.fit(x_train,y_train,epochs=2,,batch_size=100,validation_data=(x_test,y_test))
```

```python
# 验证，返回loss和metric值
model.evaluate(x=x_test,y=y_test)
```

- keras.Sequential只能构建层堆积模型
- keras.Model函数接口可构建其他复杂模型

```python
train_x = np.random.random((1000, 32))
train_y = np.random.random((1000, 10))

val_x = np.random.random((200, 32))
val_y = np.random.random((200, 10))
```

```python
# 通过keras.Model函数接口构建全连接网络
inputs = keras.Input(shape=(32,)) # 返回输入占位符
x1 = layers.Dense(64,activation='relu')(inputs)
x2 = layers.Dense(64,activation='relu')(x1)
outputs = layers.Dense(10,activation='softmax')(x2)
model = keras.Model(inputs,outputs)
model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
model.fit(train_x,train_y,batch_size=20,epochs=10,validation_data=(val_x,val_y))
```

- Model类继承

```python
class ModelTest(keras.Model):
    def __init__(self,input_nums,output_nums):
        super(ModelTest, self).__init__(name='ModelTest')
        # 定义层
        self.x = layers.Dense(64,input_shape=(32,))
        self.x1 = layers.Dense(128,activation='relu')
        self.x2 = layers.Dense(64,activation='relu')
        self.x3 = layers.Dense(output_nums,activation='softmax')
    def call(self,inputs):
        # 定义前向传递
        i = self.x(inputs)
        x1 = self.x1(i)
        x2 = self.x2(x1)
        return   self.x3(x2)
```

```python
model_t = ModelTest(32,10)
model_t.compile(optimizer=keras.optimizers.Adam(0.01),loss=keras.losses.categorical_crossentropy,metrics=['accuracy'])
model_t.fit(x=train_x,y=train_y,epochs=10,batch_size=100)
```
